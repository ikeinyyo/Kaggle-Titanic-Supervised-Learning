```python
def process_age(df,cut_points,label_names):
    df["Age"] = df["Age"].fillna(-0.5)
    df["Age_categories"] = pd.cut(df["Age"],cut_points,labels=label_names)
    return df

cut_points = [-1,0,18,100]
label_names = ["Missing","Child","Adult"]

train = process_age(train,cut_points,label_names)
test = process_age(test,cut_points,label_names)
```

```python
def create_dummies(df,column_name):
    dummies = pd.get_dummies(df[column_name],prefix=column_name)
    df = pd.concat([df,dummies],axis=1)
    return df
```

```python
def plot_confusion_matrix(y_true, y_pred, classes,
                          normalize=False,
                          title=None,
                          cmap=plt.cm.Blues):
    cm = confusion_matrix(y_true, y_pred)
  
    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation='nearest', cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # We want to show all ticks...
    ax.set(xticks=np.arange(cm.shape[1]),
           yticks=np.arange(cm.shape[0]),
           title=title,
           ylabel='True label',
           xlabel='Predicted label')

    # Rotate the tick labels and set their alignment.
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
             rotation_mode="anchor")

    # Loop over data dimensions and create text annotations.
    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, format(cm[i, j], fmt),
                    ha="center", va="center",
                    color="white" if cm[i, j] > thresh else "black")
    fig.tight_layout()
    return ax
```

```python
from sklearn.metrics import classification_report

def show_data(lr):
    lr.fit(train[columns], train['Survived'])
    predictions = lr.predict(test_X)
    accuracy = accuracy_score(test_y, predictions)

    print(accuracy) 
    print(lr.n_iter_)

    plot_confusion_matrix(test_y, predictions, ['Survived', 'Died'], title="Test data")
    plt.show()
    print(classification_report(test_y, predictions, target_names=['Survived', 'Died']))

    predictions = lr.predict(train_X)
    plot_confusion_matrix(train_y, predictions, ['Survived', 'Died'], title="Train data")
    plt.show()
    print(classification_report(train_y, predictions, target_names=['Survived', 'Died']))
```

```python
lr = LogisticRegression()
show_data(lr)
```

```python
lr = LogisticRegression(random_state=0, penalty="l2", max_iter=1000, tol=0.001, solver='liblinear', multi_class='ovr')
show_data(lr)
```

```python
lr = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial')
show_data(lr)
```
